## di gabungkan dengan membuat kolom baru
* agar bisa di group by

```python
nama = lis
id = np.array([random.choice(range(1, 2)) for i in range(18309)])

e = pd.DataFrame({'id_nama' : id, 'nama_lis' : nama})
e
     
```

```python
tes_id = np.array([random.choice(range(1, 2)) for i in range(1000)])

tes_id

# membuat id untuk groupby
# dengan random coise seharusnya nilainya ada dua pilahan antara 1 dan 2
# tapi kenapa ini satu semua ya
# ini bisa di jadikan id untuk perbandingan lis
# seharusnya ada cara lain untuk perbandingan id 
```

```python
split_bingung_juga_masih_belum_paham = e.groupby('nama_lis', as_index=False)['id_nama'].sum()
split_bingung_juga_masih_belum_paham.sort_values('id_nama', ascending=False)

# dengan nilai "id_nama" adalah 1 = bisa di jumlahkan untuk menghasilkan julmah lis
```

```python
plt.rcParams["figure.figsize"] = (7,5)

sns.barplot('nama_lis', 'id_nama', data = split_bingung_juga_masih_belum_paham.sort_values('id_nama', ascending=False).head(5))
```
